```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
go1
```

```python
go1 = pd.read_csv("./pe/GO/GO_traj_0_sub_topGO.csv", sep = ",").iloc[:7]
func = go1['Term']
# func = [x.split("~")[1] for x in func]
pval = go1['weightFisher']
pval = -np.log(pval)

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot()
barWidth = 0.7
go = ax.barh(np.arange(len(func)), pval.values, height = barWidth, capsize = 7, color = 'gray')

ax.set_xlabel("-log2(Pvalue)", fontsize = 15)
plt.yticks(np.arange(len(func)), func, fontsize = 15)
plt.gca().invert_yaxis()
ax.set_title("DE genes on Path 0 (cycle part)", fontsize = 20)
fig.savefig(fname = "./pe/plots/GO_path0.png", bbox_inches = 'tight')
```
